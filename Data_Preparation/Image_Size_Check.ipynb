```python
from PIL import Image
import os
import pandas as pd

image_folder_path = "C:\\Users\\grees\\OneDrive\\Masters\\Spring 2024\\CIS_630_Advanced_Artifical_Intelligence_Systems\\Project\\archive (1)\\test\\surprise"  # Replace with the actual path to your images folder

image_info_list = []

for filename in os.listdir(image_folder_path):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
        image_path = os.path.join(image_folder_path, filename)

        # Open the image
        with Image.open(image_path) as img:
            # Extracting image information
            image_info = {
                "filename": filename,
                "width": img.width,
                "height": img.height,
            }

            image_info_list.append(image_info)

df = pd.DataFrame(image_info_list)

excel_file_path = "C:\\Users\\grees\\Desktop\\Output\\Test\\surprise.xlsx"
df.to_excel(excel_file_path, index=False)

print(f"Image information saved to {excel_file_path}")
```
